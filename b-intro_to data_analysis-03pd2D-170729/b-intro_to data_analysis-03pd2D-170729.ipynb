```python
grades_df = pd.DataFrame(
    data={'exam1': [43, 81, 78, 75, 89, 70, 91, 65, 98, 87],
          'exam2': [24, 63, 56, 56, 67, 51, 79, 46, 72, 60]},
    index=['Andre', 'Barry', 'Chris', 'Dan', 'Emilio', 
           'Fred', 'Greta', 'Humbert', 'Ivan', 'James']
)
```

```python
grades_df
```

```python
def standardized_column(column):
    return (column - column.mean()) / column.std()
```

```python
standardized_column(grades_df["exam1"])
```

```python
 def convert_grades_curve(exam_grades):
    return exam_grades.apply(standardized_column)
```

```python
convert_grades_curve(grades_df)
```

```python
df = pd.DataFrame({
    'a': [4, 5, 3, 1, 2],
    'b': [20, 10, 40, 50, 30],
    'c': [25, 20, 5, 15, 10]
})
```

```python
df['a'].sort_values(ascending=False)[3]
```

```python
def find_sencond_big(column):
    sorted_value = column.sort_values(ascending=False)
    return sorted_value.iloc[1]
    
```

```python
find_sencond_big(df['a'])
```

```python
def second_largest(df):
    
    return df.apply(find_sencond_big)
```

```python
second_largest(df)
```
